```python
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore

# Read the CSV file
df = pd.read_csv('A_etoile_LCS_performance.csv')

# Plot the data
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['String Length'], df[column], label=column)

plt.xlabel('String Length')
plt.ylabel('Performance')
plt.title('A* LCS Performance')
plt.legend()
plt.grid(True)
plt.show()
```
