```python
import matplotlib.pyplot as plt
import pandas as pd

# Function to read data from CSV files and plot execution times
def plot_lcs_execution_time(c_file, cpp_file):
    c_data = pd.read_csv(c_file)
    cpp_data = pd.read_csv(cpp_file)

    plt.figure(figsize=(10, 6))

    plt.plot(c_data['i'], c_data['time'], label='C', color='b', marker='o')

    plt.plot(cpp_data['i'], cpp_data['time'], label='C++', color='r', marker='x')

    plt.xlabel('String Length (i)', fontsize=12)
    plt.ylabel('Execution Time (seconds)', fontsize=12)
    plt.title('LCS Execution Time Comparison: C vs C++', fontsize=14)
    
    plt.legend()

    plt.xscale('log')
    plt.yscale('log')

    plt.savefig('lcs_execution_time_comparison.png')

    plt.show()

```

```python
c_file = 'lcs_results.csv'  
cpp_file = 'lcs_cpp_results.csv' 
plot_lcs_execution_time(c_file, cpp_file)
```
